```python
from PIL import Image
import imagehash
```

```python
hash0 = imagehash.average_hash(Image.open('test_photos/test5photo0.png')) 
hash1 = imagehash.average_hash(Image.open('test_photos/test5photo6.png')) 
print(int(hash0),int(hash1))
cutoff = 5  # maximum bits that could be different between the hashes. 

if hash0 - hash1 < cutoff:
  print('images are similar')
else:
  print('images are not similar')
```

```python
c=1
for i in range(1,9):
    hash0 = imagehash.average_hash(Image.open(f'test_photos/test5photo{i-1}.png')) 
    hash1 = imagehash.average_hash(Image.open(f'test_photos/test5photo{i}.png')) 
    cutoff=5
    if hash0 - hash1 < cutoff:
        print('images are similar')
    else:
        print('images are not similar')
        c+=1
print(c)
```
